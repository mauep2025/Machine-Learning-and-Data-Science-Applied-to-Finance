#### 1. Escribe una función que acepte un string y calcule el número de mayúsculas y minúsculas en él.  ( 1 )

```python
# 1) Se pide la palabra o la oración tipo String

text = str(input ("Introduce un texto : ")) 

# Definimos las minusculas y las mayusculas

mayu= "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"

minu= "abcdefghijklmnñopqrstuvwxyz"

low_c = 0
upper_c = 0

# Contamos y clasificamos

for s in text:
    if s in mayu:
       upper_c+=1
    elif s in minu:
       low_c +=1
    
print("El número de minusculas es:", low_c)
print("El número de mayuscula es:", upper_c)



```

#### 2. Escribe un programa que calcule el `factorial` de un número entre 0 y 100 ingresado al programa por el usuario. Si el programa recibe como valor un número fuera del rango deberá volver a solicitar que se ingrese el número. (1.5)

```python

# Definimos el factorial 
def factorial2(num):
    r = 1  
    if num < 0:        
        return print("No hay factorial para numeros negativos, vuelva a introducir otro número")
        
    if num >= 100:
        return print("No se admiten valores mayores al 100. Vuelva a introducir un nuevo número menor a 100 por favor:")
    for d in range(2, num + 1):
        r *= d
    return r    


# Se introduce el número entre 0 y 99

numero = int(input ("Elige algun número para el factorial: "))
result= factorial2(numero)
print(result)

```
