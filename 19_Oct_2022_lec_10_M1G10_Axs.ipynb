```python
import pandas as pd, numpy as np
```

```python
df = pd.read_csv('employees_2.csv')
df.head()
```

```python
# calcular un incremento en el bono (Bonus %) basado en las siguientes condiciones:
# si el salario es menor a 90000 usd Y el equipo es de Marketing --> incremento 10%
# si el salrio es mayor o igual a 90000 Y el quipo es de Finanzas --> incremento 11.5%
# e.o.c.: dejamos el bono igual:

def new_bonus(row):
    if row['Salary'] < 90000 and row['Team']== 'Marketing':
        return row['Bonus %']*1.1
    elif row['Salary'] >=90000 and row['Team']== 'Finance':
        return row['Bonus %']*1.115
    else:
        return row['Bonus %']

```

```python
df.apply(new_bonus, axis = 'columns')
```

```python
df['New bonus']= df.apply(new_bonus, axis = 'columns')
df.head()
```

## Cómo iterar sobre un objeto Groupby

```python
# Sup que quisiéramos un dataframe con el salario + alto de c/ equipo.
equipos.get_group('Finance')
```

```python
df_aux = pd.DataFrame(columns = df.columns)
df_aux
```

```python
for equipo, dataf in equipos:
    highest_group_salary = dataf.nlargest(1, 'Salary')
    df_aux = df_aux.append(highest_group_salary)
    
```

## Numpy
### Numeric Python ---> librería que se utiliza principalmente para trabajar con arreglos de números. También cuenta con funciones que permiten trabajar Álgebra Lineal y Matrices.

```python
import numpy as np
```

```python
# Definamos una matriz de 2 renglones por 3 columnas
arr1 = np.array([[1,2,3],[4,5,6]])
arr1
```

Una de las principales ventajas de usar numpy es el poder usar vectorización de funciones aplicadas a un np.array. o.e. Poder aplicar una función a cada arreglo sin la necesidad de usar un for loop, lo cual lo hace más eficiente.

```python
%%timeit -n 5
sum([i**2 for i in range(1000000)])
```

```python
%%timeit -n 5
(np.arange(1000000)**2).sum()
```
