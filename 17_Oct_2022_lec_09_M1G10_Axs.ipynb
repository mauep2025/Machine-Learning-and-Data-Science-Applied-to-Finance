### Trabajo en el directorio actual
#### El directorio actual se llama `Current Working Directory`, y puede ser interpretado como una `carpeta` en la que python está operando.  Cada vez que utilizamos sólo el nombre de un archivo, python asume que se trata de un archivo en el CWD. 

```python
# Para conocer el CWD se utliza el método getcwd()
```

```python
import os
#utilizamos elmétodo getcwd() de os
cwd = os.getcwd()
print("la carpeta em la que estás trabajando es: ", cwd)
```

```python
#Podemos cambiar el directorio sobre el cual se trabaja --> usamos el método chdir()

# Como ya importamos os, podemos utlizar el método.
#Primero definamos una función que nos de la carpeta actual.

def directorio_actual():
    print("La carpeta en la que estás trabajando es: ")
    print(os.getcwd())
    print()
```

```python
directorio_actual()
```

```python
os.chdir('../')
directorio_actual()
```

```python
# Aquí podemos utilizar el método listdir() para poder ver el direcotio completo del CWD.
os.listdir()
```

```python
# Sup queremos volver al directorio:
os.chdir('/Users/alexiscaballero/Documents/DIPLOMADO ITAM DS & ML/modulo 1 G10 comienza 19sep2022')
```

```python
#revisar si ya estamos en la carpeta original:
directorio_actual()
```

### Creación de carpetas
#### Para crear una carpeta se pueden utilizar los métodos `mkdir() y makedirs()` de  os

```python
# El uso de makedir()  crea una carpeta EN EL PATH DESCRITO --> cuando ya tenemos el directorio

carpeta = 'Data ITAM prueba 17oct2022'

# Definimos la ruta en donde queremos colocarla

camino = '/Users/alexiscaballero/Documents/DIPLOMADO ITAM DS & ML/modulo 1 G10 comienza 19sep2022' #esta ruta ya existe

# unimos carpeta y 'camino' a través del método join de path:

path = os.path.join(camino, carpeta)

#creamos el directorio

os.mkdir(path)

print('Se creó la carpeta', carpeta)

os.listdir('/Users/alexiscaballero/Documents/DIPLOMADO ITAM DS & ML/modulo 1 G10 comienza 19sep2022')




```

```python
# El uso de makedirs() crea carpetas de forma RECURSIVA
# eSTO significa que cuando crea la última carpeta, si es necesario tmb creal las carpetas de la ruta que no existían

#determinamos la carpeta a crear
carpeta = 'DS prueba 17oct'

#determinamos el camino / ruta en donde quiera colocarla:

camino = '/Users/alexiscaballero/Documents/DIPLOMADO ITAM DS & ML/modulo 1 G10 comienza 19sep2022/prueba' #esta no existe

#unimos ruta y directorio a través de join:

path = os.path.join(camino, carpeta)

#creamos el directorio

os.makedirs(path)

```

```python
os.listdir('/Users/alexiscaballero/Documents/DIPLOMADO ITAM DS & ML/modulo 1 G10 comienza 19sep2022/')
```

```python
os.listdir('/Users/alexiscaballero/Documents/DIPLOMADO ITAM DS & ML/modulo 1 G10 comienza 19sep2022/prueba')
```

### Eliminación de carpetas
#### Para eliminar directorios o archivos específicos se utilizan los métodos `remove()` y `rmdir()`

```python
# método remove()
# usa para quitar un archivo en una ruta
# NO puede eliminar una carpeta.

archivo = 'archivo_a_eliminar.rtf'

# determinamos la ubicación del archivo

camino = '/Users/alexiscaballero/Documents/DIPLOMADO ITAM DS & ML/modulo 1 G10 comienza 19sep2022'

path = os.path.join(camino, archivo)
```

```python
os.listdir('/Users/alexiscaballero/Documents/DIPLOMADO ITAM DS & ML/modulo 1 G10 comienza 19sep2022/')
```

```python
#ahora eliminamos el archivo

os.remove(path)
```

```python
#el método rmdir() sirve para eliminar una carpeta que se encuentra vacía

carpeta = 'Data ITAM prueba 17oct2022'
camino = '/Users/alexiscaballero/Documents/DIPLOMADO ITAM DS & ML/modulo 1 G10 comienza 19sep2022/'
path = os.path.join(camino, carpeta)

```

```python
# quitamos la carpeta
os.rmdir(path)
```

```python
os.listdir('/Users/alexiscaballero/Documents/DIPLOMADO ITAM DS & ML/modulo 1 G10 comienza 19sep2022/')
```

## metodo where()

```python
df = pd.read_csv('employees_2.csv')
df.head()
```

```python
is_Sarah = df['First Name'] == 'Sarah'
df[is_Sarah]
```

```python
df.where(is_Sarah)
```

```python
[i for i in is_Sarah.index if is_Sarah[i]]
```

```python
df.where(is_Sarah).iloc[[2,4,5,7,224, 263, 359, 404, 417, 502, 695, 955, 977,988]]
```

```python
df.where(is_Sarah).iloc[[i for i in is_Sarah.index if is_Sarah[i]]]
```
