```python
#7. Utilizando los valores de arriba, contruye una función que determine si el salario es:
# 'bajo' si salario <= al percentil 25
# 'medio' si salario esta entre el percentil 25 (excluyendo a este) y el percentil 75 (incluyéndolo)
# 'alto' si el salario es mayor al pcntil 75.
```

```python
def clasifica_salario(salario):
    if salario <= 55292.25:
        return 'bajo'
    elif salario <= 126997.75:
        return 'medio'
    else:
        return 'alto'
```

```python
emp['salary'].apply(clasifica_salario)
#serie.apply(función)
#emp.rang_salario = emp['salary'].apply(clasifica_salario)
```
