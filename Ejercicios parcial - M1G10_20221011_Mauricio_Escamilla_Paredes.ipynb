###  <span style='font-family: Arial'> <font color=cornflowerblue> ***1. Escribe un programa que imprima las tablas de multiplicar del 1 al 10.*** </font> <font color=blue> (1.5 pts)</font>
##### <font color=red> Hint:</font> puedes usar un ciclo 'for'

```python
# Se llaman a las librerias
import pandas as pd
import numpy as np

# Se usa un doble "for" para realizar las operaciones y mostrar los reultados
print("Tablas de Multiplicar del 1 al 10")
for j in range (1,11):
    print(j, end="\t")
print()
print("__________________________________________________________________________\n")  
for m in range (1,11):
    for i in range (1,11):
        print(m*i, end="\t")
    print("\n")  

```

```python
import numpy as np
# 7 )Se muestra el renglón (fila) con el salario mas alto
  # Primero localizamos el renglon con posición del valor máximo

print("7) Fila con el empleado con salario mas alto es:")
fila = emp_data['salary'].idxmax()
print(fila)
emp_data.loc[[fila]]


```

```python
# 8 ) Mujeres con salario mayor a la media
#print(" La mediana del salario es:")

print("8) Mujeres con salario mayor a la mediana:")

mmm = emp_data['salary'].median()

emp_w_data = emp_data[(emp_data["gender"] == 'F') & (emp_data["salary"] > mmm) ]
#emp_w_data_1 = emp_data[emp_data["salary"] > 89686.500000]
emp_w_data
#emp_w_data_1
```

```python
# 9 ) función "above_media"
print("9) Función Salario mayor a la media")


def above_media(salario_3):
    if salario_3 > mmm:
        return 'True: Salario mayor a la mediana'
    elif salario_3 <= mmm:
        return 'False: Salario menor a la mediana'
    

```
