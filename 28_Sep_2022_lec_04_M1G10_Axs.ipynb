## Método rank()


```python
## opera sobre UNA SERIE. i.e. una columna de nuestro df
# necesitamos que no haya nulos
# necesitamos que el dato sea entero
```

```python
nba = pd.read_csv('nba.csv').dropna(how = 'all')
nba.Salary = nba.Salary.fillna(0).astype('int')
nba.head(3)
```

```python
#7. Utilizando los valores de arriba, contruye una función que determine si el salario es:
# 'bajo' si salario <= al percentil 25
# 'medio' si salario esta entre el percentil 25 (excluyendo a este) y el percentil 75 (incluyéndolo)
# 'alto' si el salario es mayor al pcntil 75.
```

```python
def clasifica_salario(salario):
    if salario <= 55292.25:
        return 'bajo'
    elif salario <= 126997.75:
        return 'medio'
    else:
        return 'alto'
```

```python
#8 Aplica la función anterior a la columna salary de la tabla employees:
emp['salary'].apply(clasifica_salario)
```
